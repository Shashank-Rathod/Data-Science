```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Generate months
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Generate random sales data (1000-5000 units per month)
sales_A = np.random.randint(1000, 5000, size=12)
sales_B = np.random.randint(800, 4500, size=12)
sales_C = np.random.randint(500, 4000, size=12)

# Create DataFrame
df = pd.DataFrame({
    "Month": months,
    "Sales_A": sales_A,
    "Sales_B": sales_B,
    "Sales_C": sales_C
})

# Save as CSV
df.to_csv("sales_data.csv", index=False)

print("Dataset created and saved as 'sales_data.csv'")
print(df)
```

```python
# Load dataset
df = pd.read_csv("sales_data.csv")

# Display the first few rows
print(df.head())

# Summary statistics
print(df.describe())

# Find the month with the highest sales for each product
max_sales_A = df.loc[df["Sales_A"].idxmax(), "Month"]
max_sales_B = df.loc[df["Sales_B"].idxmax(), "Month"]
max_sales_C = df.loc[df["Sales_C"].idxmax(), "Month"]

print(f"Highest sales for Product A: {max_sales_A}")
print(f"Highest sales for Product B: {max_sales_B}")
print(f"Highest sales for Product C: {max_sales_C}")
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(10, 5))

# Plot sales data
plt.plot(df["Month"], df["Sales_A"], marker='o', linestyle='-', color='blue', label="Product A")
plt.plot(df["Month"], df["Sales_B"], marker='s', linestyle='--', color='red', label="Product B")
plt.plot(df["Month"], df["Sales_C"], marker='d', linestyle=':', color='green', label="Product C")

# Formatting
plt.xlabel("Month")
plt.ylabel("Sales (Units)")
plt.title("Monthly Sales Data")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

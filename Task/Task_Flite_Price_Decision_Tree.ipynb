```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv(r"D:\Data Science\science\Project\Flight_Price_Dataset.csv")
df
```

```python
df = df.drop(['Unnamed: 0', 'flight'], axis=1)
```

```python
x = df.drop(columns=['flight', 'price'])
y = df['price']
```

```python
label_encoders = {}
categorical_cols = ['airline', 'source_city', 'departure_time', 'arrival_time', 'destination_city', 'class']
```

```python
from sklearn.preprocessing import LabelEncoder
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
# x = pd.get_dummies(data=x, columns=['airline','source_city','departure_time','stops',
# 'arrival_time','destination_city','class'], drop_first=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

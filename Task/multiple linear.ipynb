```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import pickle
```

```python
df = pd.read_csv('multi_regression_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns = 'Monthly_Revenue')
y = df['Monthly_Revenue']
```

```python
x.head()
```

```python
y.head()
```

```python
df['Employee_Count'].unique()
```

```python
x = pd.get_dummies(x, columns = ['Employee_Count'], drop_first = True)
x.head()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

x[['Marketing_Spend', 'Customer_Satisfaction']] = scaler.fit_transform(x[['Marketing_Spend', 'Customer_Satisfaction']])
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=102)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(x_train,y_train)
model
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
train_acc = lr.predict(x_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
